## Plot histograms and pdfs
For each of the gamma, normal, and beta distributions, use Python to do the following:
1. On the same set of axes, plot the distribution for 2 different values of its parameters.
2. Generate 100 samples from the distribution for one setting of its parameters and plot a histogram of the sample values. On the same plot, compare the histogram to the probability density function. The histogram will be noisy, but should align fairly well with the pdf.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

### Gamma distribution

```python
plt.figure(figsize=(8, 4))

x = np.linspace(0, 5, 200) 
alpha = 1
beta = 1

plt.plot(x, stats.gamma.pdf(x, a=alpha, scale=1/beta), label='default')
plt.plot(x, stats.gamma.pdf(x, a=2, scale=1/beta), label='alpha = 2')
plt.plot(x, stats.gamma.pdf(x, a=alpha, scale=1/2), label='beta = 2')

plt.xlim(-0.5, 5)
plt.legend() 
plt.show()
```

```python
plt.figure(figsize=(8, 4))

x = np.linspace(0, 5, 200) 
alpha = 2
beta = 2

samples = stats.gamma.rvs(alpha, scale=1/beta, size=1000)

plt.plot(x, stats.gamma.pdf(x, a=alpha, scale=1/beta), label='distribution')
plt.hist(samples, density=True, bins=50, alpha=0.5, label='samples')

plt.legend()
plt.show()
```

### Normal

```python
plt.figure(figsize=(8, 4))

x = np.linspace(-6, 6, 200) 
mu = 0
sigma = 1

plt.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma), label='default')
plt.plot(x, stats.norm.pdf(x, loc=3, scale=sigma), label='mu = 3')
plt.plot(x, stats.norm.pdf(x, loc=-3, scale=sigma), label='mu = -3')
plt.plot(x, stats.norm.pdf(x, loc=mu, scale=0.3), label='sigma = 1/3')

plt.xlim(-6,6)
plt.legend() 
plt.show()
```

```python
plt.figure(figsize=(8, 4))

x = np.linspace(-3, 3, 200) 
mu = 0
sigma = 0.5

samples = stats.norm.rvs(mu, sigma, size=1000)

plt.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma), label='distribution')
plt.hist(samples, density=True, bins=50, alpha=0.5, label='samples')

plt.legend()
plt.show()
```

### Beta

```python
plt.figure(figsize=(8, 4))

x = np.linspace(0, 1, 200) 
alpha = 1
beta = 1

plt.plot(x, stats.beta.pdf(x, a=alpha, b=beta), label='default')
plt.plot(x, stats.beta.pdf(x, a=4, b=beta), label='alpha = 3')
plt.plot(x, stats.beta.pdf(x, a=8, b=beta), label='alpha = 8')

plt.plot(x, stats.beta.pdf(x, a=alpha, b=4), label='beta = 3')
plt.plot(x, stats.beta.pdf(x, a=alpha, b=8), label='beta = 8')

plt.plot(x, stats.beta.pdf(x, a=8, b=8), label='alpha = beta = 8')


plt.xlim(0, 1)
plt.legend() 
plt.show()
```

```python
plt.figure(figsize=(8, 4))

x = np.linspace(0, 1, 200) 
alpha = 2
beta = 10

samples = stats.beta.rvs(alpha, beta, size=1000)

plt.plot(x, stats.beta.pdf(x, a=alpha, b=beta), label='default')
plt.hist(samples, density=True, bins=50, alpha=0.5, label='samples')

plt.legend()
plt.show()
```
